```python
app_genres  = app_data.drop_duplicates(subset=["App"], keep='last')
print(app_genres.loc[:, 'Size'].unique())
```

Filtro aquellos que tengan Varies with device

```python
filtered_genres = app_genres.loc[~app_genres["Size"].str.contains("Varies"), :]
```

```python
# ----ATENCION---- Revise esto con hige y concluimos en dejarlo en el formato este, sin assign
filtered_genres['formated_genre'] = filtered_genres.Genres.map(lambda x : [genre.lower().strip() for genre in x.split(";")])

# Esto es lo que conlcuimos NO usar
# filtered_genres = filtered_genres.assign(formated_genre= filtered_genres.Genres.map(lambda x : [genre.lower().strip() for genre in x.split(";")]))
```

```python
def formatear_espacio_a_bytes(tamanio):
    if('k' in tamanio):
        return float(tamanio.strip('k ')) * 1024
    elif('M' in tamanio):
      return float(tamanio.strip('M ')) * 1024 * 1024
    return float(tamanio.strip('+ ').replace(',', '.'))

filtered_genres["size_in_bytes"] = filtered_genres.Size.map(formatear_espacio_a_bytes)
```

Utilizo el explode de la misma forma que lo usamos en clase (con Hige), para hacer una fila por cada genero (creo que debe hacer un .split por dentro y despues hace lo otro)

```python
formated_genres = filtered_genres.explode('formated_genre')
formated_genres.formated_genre.unique()
```

```python
formated_genres = formated_genres.loc[~formated_genres["formated_genre"].str.contains("february"), :] # Elimino el campo invalido
```

El tamaño está en bytes, mas abajo dejo en MB redondeados

```python
size_by_genre = formated_genres.groupby('formated_genre').agg({'size_in_bytes': 'sum'})
size_by_genre
#Este es el tamaño en bytes
```

Redondeo los MB para que se aprecie mejor, esto es meramente demostrativo

```python
size_by_genre["size_in_mb"] = size_by_genre.size_in_bytes.map(lambda x: round(x/(1024*1024)))
size_by_genre.sort_values(by="size_in_mb")
```
