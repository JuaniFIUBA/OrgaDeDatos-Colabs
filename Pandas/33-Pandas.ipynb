```python
app_data = pd.read_csv('/content/drive/MyDrive/aa-OrgaDeDatos/googleplaystore.csv')
app_reviews = pd.read_csv('/content/drive/MyDrive/aa-OrgaDeDatos/googleplaystore_user_reviews.csv')
```

Elimino las filas cuyos valores son repetidos, o no tiene sentido que sean na (ya se que los tienen por analisis previos)


```python
app_reviews_filtered = app_reviews.dropna(subset=["Sentiment_Subjectivity"]).drop_duplicates(keep='last')
app_data_filtered = app_data.drop_duplicates(subset=["App"], keep='last')
```

Voy a juntar los dataframes con un merge inner en base a la columna App, para posteriormente agruparlos y calcularles el promedio.

```python
categories_by_sentiment = app_data_filtered.merge(app_reviews_filtered, on='App')
mean_sentiment_by_cat = categories_by_sentiment.groupby('Category').agg({'Sentiment_Subjectivity': 'mean'})
# Esto es para que tenga sentido y no diga Sentiment_Subjectivity
mean_sentiment_by_cat.rename(columns={'Sentiment_Subjectivity': 'Sentiment_Subjectivity_Mean'}, inplace=True)
mean_sentiment_by_cat.sort_values(by='Sentiment_Subjectivity_Mean', inplace=True, ascending=False)
mean_sentiment_by_cat
```
